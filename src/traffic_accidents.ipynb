9,shows the covariance between each pair of the columns

```python
import numpy as np

df_selected = df.select_dtypes(include=[np.number])
df_selected = df_selected.dropna()
cov_matrix = df_selected.cov()
cov_matrix
```

From the table, LONGITUDE and LATITUDE has the highest correlation. However, the values for LONGITUDE are from 73 to 74 and the values for LATITUDE are from 40 to 41. The values are not vary too much which mean the error will be very small. Therefore the correlation is very high.
The columns that are really correlated are NUMBER OF PERSONS INJURED, NUMBER OF PERSONS KILLED, NUMBER OF PEDESTRIANS INJURED, NUMBER OF PEDESTRIANS KILLED, NUMBER OF CYCLIST INJURED, NUMBER OF CYCLIST KILLED, NUMBER OF MOTORIST INJURED, NUMBER OF MOTORIST KILLED. The reason is common sense, more people die in a car accident means more people are injured.

10,which month did the most number of accidents occur?

I created a new column called MONTH by split the crash date column and take the first piece and called value_counts on it. Also rename the x axis using rename and a lambda function

```python
import calendar
df['MONTH'] = df['CRASH DATE'].apply(lambda x: x.split('/')[0])
df['MONTH'].value_counts().sort_index().rename(index=lambda x: calendar.month_name[int(x)]).plot.bar()
```

From the graph, we can tell January is the month with most number of accidents
